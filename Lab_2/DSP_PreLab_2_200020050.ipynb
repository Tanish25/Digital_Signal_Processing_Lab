```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal
import scipy as sp
%matplotlib inline
```

```python
#construct and plot a DT sinosouid of frequency w = pi/6 rad/sec and phase phi = 0
file = "sine.dat"
file1 = open(file, "w")

fs=100#sampling frequency
#parameters
omega = np.pi/6 #frequency
phi = 0 #phase difference
period = 2*(np.pi/omega) #time period



n = np.linspace(0,40,fs)#to generate discrete values in sorted order
DTsin = np.sin(omega*n + phi)#discrete time sinusoid
for i in range(len(DTsin)):
  file1.write(str(DTsin[i]))
  file1.write(str("\n"))

file1.close()
#plotting the sinosoid wrt time
plt.grid()#to show grid
plt.stem(n, DTsin)
plt.xlabel('TIME')
plt.ylabel('AMPLITUDE')
plt.show()#to plot individual graphs
```

```python
#DT Square Wave
fs = 100 #sampling frequency to view the discrete samples
t = np.linspace(0,10,fs)
file = "square.dat"
file1 = open(file, "w")

DTsquare = (1+signal.square(2 * np.pi * 0.5 * t, duty=0.5)/2) - 0.5#discrete time square
plt.grid()
#stem used to plot discrete signal
plt.stem(t, DTsquare)# both frequency and duty cycle is 0.5

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Discrete time Square wave signal')
plt.show()
for i in range(len(DTsquare)):
  file1.write(str(DTsquare[i]))
  file1.write(str("\n"))

file1.close()
```

```python
#Discrete time triangle wave signal

#parameters
width = 0.5
f = 0.25
fs = 20#sampling frequency

file = "triangle.dat"
file1 = open(file, "w")

n = np.linspace(0,20,fs)
sawt = signal.sawtooth(np.pi*n*f,width)
triangle = (sawt + 1)/2
for i in range(len(triangle)):
  file1.write(str(triangle[i]))
  file1.write(str("\n"))

file1.close()
#plots
plt.grid()
plt.stem(n,triangle)#stem used to discretise the CT signal
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Discrete time triangle wave signal')
plt.show()
```
